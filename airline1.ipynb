```python
import gradio as gr

# prediction function
def predict_delay(carrier, airport, weather_ct, late_aircraft_ct):
    try:
        # Encode categorical inputs
        carrier_encoded = obj1.transform([carrier])[0]
        airport_encoded = obj2.transform([airport])[0]
    except ValueError:
        return "❌ Enter valid carrier and airport from dataset"

    # Prediction
    pred = model.predict([[carrier_encoded,
                            airport_encoded,
                            weather_ct,
                            late_aircraft_ct]])[0]

    return f"Predicted Arrival Delay: {pred:.2f} minutes"


# Gradio interface
iface = gr.Interface(
    fn=predict_delay,
    inputs=[
        gr.Textbox(label="Carrier Code (e.g., AA, DL)"),
        gr.Textbox(label="Airport Code (e.g., ATL, JFK)"),
        gr.Number(label="Weather Delay Count"),
        gr.Number(label="Late Aircraft Delay Count")
    ],
    outputs=gr.Textbox(label="Prediction Result"),
    title="✈️ Airline Delay Prediction",
    description="Enter flight details to predict arrival delay (minutes)"
)

iface.launch()
```
